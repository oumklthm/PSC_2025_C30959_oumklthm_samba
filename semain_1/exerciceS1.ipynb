```python
#exo1
def f(x):
    return x**2 - 0.25*x + 5
# Vérification pour x = 2.3
x = 2.3
resultat = f(x)
print(f"f({x}) = {resultat}")
# Vérification si c'est un zéro
if resultat == 0:
    print(f"{x} est un zéro de la fonction.")
else:
    print(f"{x} n'est pas un zéro de la fonction.")
```

```python
# exo2
import cmath 
n = 3
x = cmath.pi / 4  # pi/4

# Partie  (cos x + i sin x)^n
z = cmath.cos(x) + 1j * cmath.sin(x)
left = z ** n

# Partie : cos(nx) + i sin(nx)
right = cmath.cos(n * x) + 1j * cmath.sin(n * x)


print("Left =", left)
print("Right =", right)


if abs(left - right ) < 1e-10:
    print(" La formule de De Moivre est vérifiée.")
else:
    print(" La formule de De Moivre n'est pas vérifiée.")
```

```python
#exo3
import cmath
x = cmath.pi / 4 

# Partie e^(ix)
left = cmath.exp(1j * x)

# Partie cos(x) + i sin(x)
right = cmath.cos(x) + 1j * cmath.sin(x)
print("e^(ix)         =", left)
print("cos(x)+i·sin(x) =", right)


if abs(left - right) < 1e-10:
    print(" La formule d'Euler est vérifiée.")
else:
    print("La formule d'Euler n'est pas vérifiée.")
```

```python
#exo4
u = 1.0  # valeur de U0
uold = 10.0  # valeur pour comparer

for iteration in range(2000):
    if not abs(u - uold) > 1e-8:  # si la différence est très petite
        print("Convergence")
        break 
    uold = u
    u = 2 * u 
else:
    print("No convergence") 
```
